# 1. Perform a normality test (e.g., Shapiro–Wilk test) 
### for each feature in the dataset, separately for the three groups: Healthy, Disease, and Recovered.
### - Use a significance level (α) of 0.05.
### - Specifically compare feature1, feature2, and feature4 to check for distribution similarity between groups.
### - Visualize the comparison using appropriate plots (e.g., histograms, Q–Q plots, or box plots).

```python
group_names = ['Healthy', 'Disease', 'Recovered']
dataframes = [healthy_df, disease_df, recovered_df]
```

```python
results = []
alpha=0.05
    
for df, group_name in zip(dataframes, group_names):
    for column in df.columns:
        data = df[column]
        # Perform Shapiro-Wilk test
        statistic, p_value = stats.shapiro(data)
        
        # Determine if normally distributed
        is_normal = "Yes" if p_value > alpha else "No"
        
        results.append({
            'Group': group_name,
            'Feature': column,
            'Statistic': statistic,
            'P-Value': p_value,
            'Normal (alpha=0.05)': is_normal
        })

normality_results_df = pd.DataFrame(results)
print("Shapiro-Wilk Normality Test Results (alpha=0.05):\n", normality_results_df)
```

```python
# feature wise visualization
for feature in healthy_df.columns:
    feature_results = normality_results_df[normality_results_df['Feature'] == feature]
    print(f"\n{feature}:")
    for _, row in feature_results.iterrows():
        print(f"  {row['Group']:10s}: p-value = {row['P-Value']:.6f} - {row['Normal (alpha=0.05)']}")
```

```python
focused_features = ['Feature 1', 'Feature 2', 'Feature 4']

for feature in focused_features:
    print(f"\n{feature}:")
    print("-" * 80)
    
    for df, group_name in zip(dataframes, group_names):
        data = df[feature].dropna()
        stat, p_val = stats.shapiro(data)
        
        print(f"\n{group_name}:")
        print(f"  Mean: {data.mean():.4f}")
        print(f"  Std Dev: {data.std():.4f}")
        print(f"  Median: {data.median():.4f}")
        print(f"  Shapiro-Wilk Statistic: {stat:.6f}")
        print(f"  P-Value: {p_val:.6f}")
        print(f"  Normally Distributed: {'Yes' if p_val > 0.05 else 'No'}")
        print(f"  Skewness: {stats.skew(data):.4f}")
        print(f"  Kurtosis: {stats.kurtosis(data):.4f}")
```

#### Insights from above analysis:
Feature 1 clearly separates Disease (much higher mean) from Healthy and Recovered; Healthy and Disease are approximately normal, while Recovered slightly violates normality.  
Feature 2 shows substantial variability and non-normality for Healthy and Recovered, with Recovered having the highest mean and Disease closer to normal.  
Feature 4 is non-normal across all groups, with Disease showing higher mean, strong right skew, and heavy tails, suggesting stronger pathological signal but weaker parametric assumptions.

## 2. For feature1, feature2, and feature4, test the null hypothesis that the distributions of the selected features are equal across the groups.
- State the null and alternative hypotheses clearly.
- Use an appropriate statistical test (parametric or non-parametric, depending on normality results).

```python
all_results = []
posthoc_results = {}

for feature in focused_features:
    print(f"\n{'='*80}\n{feature}\n{'='*80}")
    
    # Check normality for each group
    normality_checks = []
    for df, group_name in zip(dataframes, group_names):
        data = df[feature].dropna()
        stat, p_value = stats.shapiro(data)
        normality_checks.append(p_value > alpha)
    
    all_normal = all(normality_checks)
    
    # Check homogeneity of variance
    data_groups = [df[feature].dropna() for df in dataframes]
    levene_stat, levene_p = stats.levene(*data_groups)
    equal_var = levene_p > alpha
    
    # Hypotheses
    print("\nH0: Distributions are EQUAL across all groups (Healthy = Disease = Recovered)")
    print("H1: At least one group has a DIFFERENT distribution")
    
    # Select and perform test
    if all_normal and equal_var:
        stat, p_value = stats.f_oneway(*data_groups)
        test_name = "One-Way ANOVA"
        test_type = "Parametric"
    else:
        stat, p_value = stats.kruskal(*data_groups)
        test_name = "Kruskal-Wallis H-Test"
        test_type = "Non-Parametric"
    
    # Results
    decision = "REJECT H0" if p_value < alpha else "FAIL TO REJECT H0"
    print(f"\nTest: {test_name} | Statistic: {stat:.4f} | p-value: {p_value:.6f}")
    print(f"Decision: {decision}")
    
    all_results.append({
        'Feature': feature,
        'Test': test_name,
        'Type': test_type,
        'Statistic': stat,
        'P-Value': p_value,
        'Decision': decision,
        'Significant': p_value < alpha
    })
    
    # Post-hoc tests if significant
    if p_value < alpha:
        print(f"\nPost-hoc Pairwise Comparisons (Bonferroni α = {alpha/3:.4f}):")
        posthoc_data = []
        
        for (df1, name1), (df2, name2) in combinations(zip(dataframes, group_names), 2):
            data1 = df1[feature].dropna()
            data2 = df2[feature].dropna()
            
            if test_type == "Parametric":
                stat, p_val = stats.ttest_ind(data1, data2)
            else:
                stat, p_val = stats.mannwhitneyu(data1, data2, alternative='two-sided')
            
            significant = "YES" if p_val < alpha/3 else "NO"
            print(f"  {name1} vs {name2}: p = {p_val:.6f} | Significant: {significant}")
            
            posthoc_data.append({
                'Comparison': f"{name1} vs {name2}",
                'P-Value': p_val,
                'Significant': significant
            })
        
        posthoc_results[feature] = pd.DataFrame(posthoc_data)

summary_df = pd.DataFrame(all_results)
print(summary_df.to_string(index=False))
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

for i, feature in enumerate(focused_features):
    result = all_results[i]
    
    # Prepare data
    combined_data = []
    group_labels = []
    for df, group_name in zip(dataframes, group_names):
        data = df[feature].dropna()
        combined_data.extend(data.values)
        group_labels.extend([group_name] * len(data))
    
    plot_df = pd.DataFrame({'Value': combined_data, 'Group': group_labels})
    
    # Box plot
    ax1 = axes[0, i]
    sns.boxplot(x='Group', y='Value', data=plot_df, ax=ax1, 
               palette=colors, hue='Group')
    sns.swarmplot(x='Group', y='Value', data=plot_df, ax=ax1, 
                 color='black', alpha=0.3, size=2)
    
    significance = "SIG" if result['Significant'] else "NOT SIG"
    color = 'red' if result['Significant'] else 'green'
    ax1.set_title(f"{feature} | {result['Test']} | p={result['P-Value']:.4f} | {significance}", 
                 fontsize=10, fontweight='bold', color=color)
    ax1.grid(True, alpha=0.3, axis='y')
    
    # Violin plot
    ax2 = axes[1, i]
    sns.violinplot(x='Group', y='Value', data=plot_df, ax=ax2, 
                  palette=colors, inner='box', hue='Group')
    
    stats_text = f"{result['Type']}\nStat: {result['Statistic']:.4f}\np: {result['P-Value']:.6f}\n{result['Decision']}"
    ax2.text(0.98, 0.98, stats_text, transform=ax2.transAxes, fontsize=8,
            verticalalignment='top', horizontalalignment='right',
            bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.9))
    ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

## 3. One-way ANOVA on All Features 
Conduct hypothesis testing to compare all features among the three groups  
(Healthy, Disease, Recovered) using One-way ANOVA at three different  
significance levels:  
- α = 0.05
- α = 0.001  
- α = 0.0001  

Report which features show statistically significant differences at each significance level.

```python
# Check ANOVA assumptions for all features
assumption_results = []
features = healthy_df.columns
for feature in features:
    print(f"{feature}:")
    
    # Check normality for each group
    normality_status = []
    for df, group_name in zip(dataframes, group_names):
        data = df[feature].dropna()
        stat, p_value = stats.shapiro(data)
        is_normal = p_value > 0.05
        normality_status.append(is_normal)
        print(f"  {group_name} - Normality: {'PASS' if is_normal else 'FAIL'} (p={p_value:.6f})")
    
    all_normal = all(normality_status)
    
    # Check homogeneity of variance
    data_groups = [df[feature].dropna() for df in dataframes]
    levene_stat, levene_p = stats.levene(*data_groups)
    equal_var = levene_p > 0.05
    
    print(f"  Variance Homogeneity: {'PASS' if equal_var else 'FAIL'} (p={levene_p:.6f})")
    
    assumptions_met = all_normal and equal_var
    print(f"> ANOVA Assumptions: {'MET' if assumptions_met else 'VIOLATED'}")
    
    assumption_results.append({
        'Feature': feature,
        'All_Normal': all_normal,
        'Equal_Var': equal_var,
        'Assumptions_Met': assumptions_met
    })
    print()

assumption_df = pd.DataFrame(assumption_results)
```

```python
# Perform One-Way ANOVA and Kruskal-Wallis for all features
anova_results = []
kruskal_results = []

for feature in features:
    data_groups = [df[feature].dropna() for df in dataframes]
    
    # Perform One-Way ANOVA (as requested in assignment)
    f_stat, anova_p = stats.f_oneway(*data_groups)
    
    # Also perform Kruskal-Wallis (statistically appropriate if assumptions violated)
    h_stat, kruskal_p = stats.kruskal(*data_groups)
    
    # Check significance at each level for ANOVA
    anova_sig_005 = "YES" if anova_p < 0.05 else "NO"
    anova_sig_0001 = "YES" if anova_p < 0.001 else "NO"
    anova_sig_00001 = "YES" if anova_p < 0.0001 else "NO"
    
    # Check significance at each level for Kruskal-Wallis
    kw_sig_005 = "YES" if kruskal_p < 0.05 else "NO"
    kw_sig_0001 = "YES" if kruskal_p < 0.001 else "NO"
    kw_sig_00001 = "YES" if kruskal_p < 0.0001 else "NO"
    
    # Check if assumptions were met
    assumptions_met = assumption_df[assumption_df['Feature'] == feature]['Assumptions_Met'].values[0]
    
    anova_results.append({
        'Feature': feature,
        'Assumptions_Met': 'YES' if assumptions_met else 'NO',
        'F-Statistic': f_stat,
        'P-Value': anova_p,
        'Sig_α=0.05': anova_sig_005,
        'Sig_α=0.001': anova_sig_0001,
        'Sig_α=0.0001': anova_sig_00001
    })
    
    kruskal_results.append({
        'Feature': feature,
        'H-Statistic': h_stat,
        'P-Value': kruskal_p,
        'Sig_α=0.05': kw_sig_005,
        'Sig_α=0.001': kw_sig_0001,
        'Sig_α=0.0001': kw_sig_00001
    })

anova_df = pd.DataFrame(anova_results)
kruskal_df = pd.DataFrame(kruskal_results)

print("ONE-WAY ANOVA RESULTS (PARAMETRIC)\n",anova_df.to_string(index=False))

print(f"\nKRUSKAL-WALLIS RESULTS (NON-PARAMETRIC)\n",kruskal_df.to_string(index=False))
```

```python
comparison = []
for i, feature in enumerate(features):
    anova_sig = anova_df.iloc[i]['Sig_α=0.05']
    kw_sig = kruskal_df.iloc[i]['Sig_α=0.05']
    agree = "YES" if anova_sig == kw_sig else "NO"
    
    comparison.append({
        'Feature': feature,
        'ANOVA_Sig': anova_sig,
        'Kruskal_Sig': kw_sig,
        'Tests_Agree': agree,
        'Assumptions_Met': anova_df.iloc[i]['Assumptions_Met']
    })

comparison_df = pd.DataFrame(comparison)
print(comparison_df.to_string(index=False))
```
